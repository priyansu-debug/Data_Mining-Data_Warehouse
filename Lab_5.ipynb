```python
# Demonstrate the following Similarity and Dissimilarity Measure using Python
# 1-Euclidean Dist
# 2-Manhattan Dist
# 3-Minkowski Dist
# 4-Cosine Similiarity
# 5-Jaccard Similarity
# 6-Pearson's Correlation
```

```python
# 1-Euclidean Dist
from math import *
pt1=[1,3,5]
pt2=[2,5,3]
sqrs=(pt1[0]-pt2[0])**2+(pt1[1]-pt2[1])**2+(pt1[2]-pt2[2])**2
euc_dist=sqrt(sqrs)
print("euclidean Dist between pt1 and pt2: ",euc_dist)
```

```python
# using linealg.norm() Method of Numpy
import numpy as np
p1=np.array((4,7,9))
p2=np.array((10,12,14))
euc_dist=np. linalg.norm(p1-p2)
print("euclidean Dist between pt1 and pt2: ",euc_dist)
```

```python
# 2-Manhattan Dist
def manhattan_distance (x,y):
  return sum(abs(a-b)for a,b in zip(x,y))
print("Manhattan Distance between two points: ",manhattan_distance([10,12,15],[8,12,20]))
```

```python
#Minkowski Distance
from math import *
from decimal import Decimal
def nth_root(value, n_root):
  root_value = 1/float(n_root)
  return round (Decimal(value) ** Decimal(root_value),3)
def minkowski_distance(x,y,p_value):
    return nth_root(sum(pow(abs(a-b),p_value) for a,b in zip(x, y)),p_value)
print ("minkowski_distance between two points:",minkowski_distance ([0,3,4,5],[7,6,3,-1],3))
```

```python
# 4-Cosine Similiarity
from math import *
def square_rooted(x):
  return round(sqrt(sum([a*a for a in x])),3)

def cosine_similarity(x,y):
  numerator=sum(a*b for a,b in zip(x,y))
  denominator=square_rooted  (x)*square_rooted  (y)
  return round (numerator/float(denominator),3)
print("Cosine Similarity between two points: ",cosine_similarity([3,45,7,2],[2,54,13,15]))
```

```python
# using Method of Numpy
import numpy as np
def cosinesim(vec1, vec2):
  vec1 = np.array(vec1) # Convert to numpy array
  vec2 = np.array(vec2)
  numerator = np.dot(vec1, vec2)
  v1norm = np.sqrt(sum(vec1**2))
  v2norm = np.sqrt(sum(vec2**2))
  c = numerator/(v1norm*v2norm)
  return c
data =[[3,45,7,2], [2,54,13, 15]]
cos = cosinesim(data[0], data[1])
print("Cosine Similarity between two points:", round(cos, 3))
```

```python
# 5-Jaccard Similarity
def jaccard_sim(s1,s2):
  intersection=len(s1 & s2)  # common elements
  union=len(s1 | s2)         # Total unique elements
  return (intersection/union)
x={1,2,3}
y={2,3,4,5}
print(jaccard_sim(x,y))
```

```python
from math import*
def jaccard_similarity(x,y):
  intersection_cardinality = len(set.intersection (*[set(x),set(y)]))
  union_cardinality = len(set.union(*[set(x),set(y)]))
  return intersection_cardinality/float(union_cardinality)
print("Jaccard Similarity between two points: ",jaccard_similarity([0,1,2,5,6],[0,2,3,5,7,9]))
```
