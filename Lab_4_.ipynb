```python
df = pd.read_csv('company_sales_data.csv')
missing_values = df.isnull().sum()
display(missing_values)
```

```python
# 3) Check for missing data & handle it (using mean imputation for numeric columns)
print("\nMissing values count:")
print(df.isnull().sum())

df.fillna(df.mean(numeric_only=True), inplace=True)  # Replace NaN with mean
print("\nData after handling missing values:")
display(df)
```

```python
# 4) Handling data errors
# num_cols = df.select_dtypes(include=np.number).columns
# df[num_cols] = df[num_cols].applymap(lambda x: 0 if x < 0 else x)
```

```python
# 5. handling numerical data using binning, scaler or scaling, normalization , discretization

# Using Binning here
df['Scale_bin'] = pd.cut(df['total_units'], bins=3, labels=['Low', 'Medium', 'High'])
display(df)
```

```python
# Using Scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df[['Scale_x']] = scaler.fit_transform(df[['total_profit']])
display(df)
```

```python
# Using Normalization
from sklearn.preprocessing import MinMaxScaler
scaler_minmax = MinMaxScaler()
df_normalized = df.copy()
num_cols = df.select_dtypes(include=np.number).columns
df_normalized[num_cols] = scaler_minmax.fit_transform(df[num_cols])
print("\nNormalized Data (0–1 range):")
display(df_normalized.head())
```

```python
# Using Discretzation
df['profit_disc_equal_width'] = pd.cut(df['total_profit'], bins=4, labels=["Q1", "Q2", "Q3", "Q4"])

# Equal-frequency discretization for 'total_profit'
df['profit_disc_equal_freq'] = pd.qcut(df['total_profit'], q=4, labels=["Q1", "Q2", "Q3", "Q4"])

print("\nDiscretization Examples (Equal-width & Equal-frequency):")
display(df[['total_profit', 'profit_disc_equal_width', 'profit_disc_equal_freq']])
```

```python
# 6. read total profit of all months and show it using a line plot where
# a) Line style dotted and colour should be blue

plt.plot(df['month_number'], df['total_profit'], linestyle='dotted', color='blue')
plt.title('Total Profit Per Month')
plt.show()
```

```python
# b) Show legend() at the lower right location

plt.plot(df['month_number'], df['total_profit'], label='Total Profit')
plt.legend(loc='lower right')
plt.show()
```

```python
# c) X label name month No.

plt.plot(df['month_number'], df['total_profit'])
plt.xlabel('Month No.')
plt.show()
```

```python
# d) y label name = sold unit no.

plt.plot(df['month_number'], df['total_profit'])
plt.ylabel('Sold unit no.')
plt.show()
```

```python
# e) add a circle marker

plt.plot(df['month_number'], df['total_profit'], marker='o')
plt.show()
```

```python
# f) Line marker colours as rid

plt.plot(df['month_number'], df['total_profit'], marker='o', markerfacecolor='red')
plt.show()
```

```python
#g) line width should be three.

plt.plot(df['month_number'], df['total_profit'], linewidth=3)
plt.show()
```

```python
# 7) Read tooth-paste sell data of each month and show it using scatter plot.
plt.figure(figsize=(8,5))
plt.scatter(df['month_number'], df['toothpaste'], color='blue', label="Toothpaste Sales")
plt.xlabel('Month Number')
plt.ylabel('Number of Units Sold')
plt.title('Toothpaste Sales Data')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 8)read face cream and face wash product cells data and show it using bar chart.

# Bar chart for Face Cream & Face Wash sales
plt.figure(figsize=(8,5))

# Plot bars side-by-side for each month
plt.bar(df['month_number'] - 0.2, df['facecream'], width=0.4, label='Face Cream Sales', color='Blue')
plt.bar(df['month_number'] + 0.2, df['facewash'], width=0.4, label='Face Wash Sales', color='Red')

plt.xlabel('Month Number')
plt.ylabel('Units Sold')
plt.title('Face Cream & Face Wash Sales Data')
plt.xticks(df['month_number'])  # Show month numbers on X-axis
plt.legend()
plt.show()
```

```python
# 9) read the total profit of each month and Show it using histogram to see the most common profit range

# Histogram for total profit
plt.figure(figsize=(8,5))
plt.hist(df['total_profit'], bins=10, color='skyblue', edgecolor='black')

plt.xlabel('Profit Range')
plt.ylabel('Frequency')
plt.title('Most Common Profit Range')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# 10) calculate total cell data for last year or each product & display using pie chart.

# Calculate total yearly sales for each product (excluding month_number and categorical columns)
total_sales = df.drop(columns=['month_number', 'profit_disc_equal_width', 'profit_disc_equal_freq']).sum()
print("Total sales for last year (each product):")
display(total_sales)  # works in Jupyter/Colab

# Pie chart
plt.figure(figsize=(8,8))
plt.pie(total_sales, labels=total_sales.index, autopct='%1.1f%%', startangle=140)
plt.title('Total Sales Distribution (Last Year)')
plt.show()
```

```python
# 11) read bathing soap facewash of all months and display is using subplot.

plt.figure(figsize=(10,6))

# Subplot 1 - Bathing Soap
plt.subplot(2,1,1)
plt.plot(df['month_number'], df['bathingsoap'], marker='o', color='blue')
plt.title('Bathing Soap Sales Data')
plt.ylabel('Units Sold')

# Subplot 2 - Face Wash
plt.subplot(2,1,2)
plt.plot(df['month_number'], df['facewash'], marker='o', color='green')
plt.title('Face Wash Sales Data')
plt.xlabel('Month Number')
plt.ylabel('Units Sold')

plt.tight_layout()
plt.show()
```

```python
# 12) read all product cell data & show it using the stack plot

plt.figure(figsize=(10,6))
plt.stackplot(df['month_number'],
              df['facecream'], df['facewash'], df['toothpaste'],
              df['bathingsoap'], df['shampoo'], df['moisturizer'],
              labels=['Face Cream','Face Wash','Toothpaste','Bathing Soap','Shampoo','Moisturizer'])
plt.legend(loc='upper left')
plt.title('All Product Sales Data')
plt.xlabel('Month Number')
plt.ylabel('Units Sold')
plt.show()
```

```python
# 13) read all product cell data and show it using the mult-line graph.

plt.figure(figsize=(10,6))
plt.plot(df['month_number'], df['facecream'], label='Face Cream')
plt.plot(df['month_number'], df['facewash'], label='Face Wash')
plt.plot(df['month_number'], df['toothpaste'], label='Toothpaste')
plt.plot(df['month_number'], df['bathingsoap'], label='Bathing Soap')
plt.plot(df['month_number'], df['shampoo'], label='Shampoo')
plt.plot(df['month_number'], df['moisturizer'], label='Moisturizer')

plt.xlabel('Month Number')
plt.ylabel('Units Sold')
plt.title('Sales Data of All Products')
plt.legend()
plt.show()
```
