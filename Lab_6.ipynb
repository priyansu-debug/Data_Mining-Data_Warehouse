```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules

# Step 1: Prepare Transaction Dataset
dataset = [
    ['bread', 'butter', 'jam', 'milk'],
    ['bread', 'butter', 'milk'],
    ['bread', 'juice', 'curd'],
    ['bread', 'milk', 'juice'],
    ['butter', 'milk', 'juice'],
]

# Convert dataset into one-hot encoded DataFrame
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_ary, columns=te.columns_)

print("Transaction DataFrame:\n", df)
```

```python
# Step 2: Apply Apriori Algorithm

# min_support = 0.5 means itemsets appearing in at least 50% of transactions
frequent_itemsets = apriori(df, min_support=0.5, use_colnames=True)

print("\nFrequent Itemsets:")
print(frequent_itemsets)
```

```python
# Step 3: Generate Association Rules
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7)

# We also want Lift, so include that
rules = rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']]

print("\nAssociation Rules (with Confidence & Lift):")
print(rules)
```

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

# Load dataset
dataset = pd.read_csv("/content/Market Basket Dataset.csv")
print(dataset.head())

# Convert to list of transactions
transactions = []
for i in range(0, dataset.shape[0]):
    row = dataset.iloc[i].dropna().tolist()
    transactions.append(row)

print("Total transactions loaded:", len(transactions))

# One-hot encode
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df = pd.DataFrame(te_ary, columns=te.columns_)

# Frequent itemsets with 1% minimum support
frequent_itemsets = apriori(df, min_support=0.01, use_colnames=True)
print("Frequent Itemsets (Top 10):")
print(frequent_itemsets.head(10))

# Association rules
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.3)
rules = rules[['antecedents','consequents','support','confidence','lift']]
rules = rules.sort_values(by='lift', ascending=False)

# Convert frozensets to strings for readability
rules['antecedents'] = rules['antecedents'].apply(lambda x: ', '.join(list(x)))
rules['consequents'] = rules['consequents'].apply(lambda x: ', '.join(list(x)))

print("\nAssociation Rules (Top 10 by Lift):")
print(rules.head(10))
```

```python
#Write a program to implement FP-Growth algorithm to find frequent pattern in the transactions.
#Use Confidence and Lift measure for generating association rule.
#Compare its efficiency with Apriori algorithm.
#Python ML libraries can be used forthisproblem.
import time
import pandas as pd
from mlxtend.frequent_patterns import fpgrowth, apriori, association_rules

# Dataset (list of transactions)
dataset = [
    ['Milk', 'Onion', 'Nutmeg', 'Kidney Beans', 'Eggs', 'Yogurt'],
    ['Dill', 'Onion', 'Nutmeg', 'Kidney Beans', 'Eggs', 'Yogurt'],
    ['Milk', 'Apple', 'Kidney Beans', 'Eggs'],
    ['Milk', 'Unicorn', 'Corn', 'Kidney Beans', 'Yogurt'],
    ['Corn', 'Onion', 'Onion', 'Kidney Beans', 'Ice cream', 'Eggs']
]

# Convert dataset to one-hot encoded DataFrame
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_ary, columns=te.columns_)
print(df)

# Apriori Algorithm
start_time = time.time()
frequent_itemsets_apriori = apriori(df, min_support=0.6, use_colnames=True)
apriori_time = time.time() - start_time
# FP-Growth Algorithm
start_time = time.time()
frequent_itemsets_fpgrowth = fpgrowth(df, min_support=0.6, use_colnames=True)
fpgrowth_time = time.time() - start_time

# Generate association rules from FP-Growth frequent itemsets
rules_fpgrowth = association_rules(frequent_itemsets_fpgrowth, metric="lift", min_threshold=1.0)

# Print FP-Growth Frequent Itemsets
print("=== FP-Growth Frequent Itemsets ===")
print(frequent_itemsets_fpgrowth)

# Print FP-Growth Association Rules
print("\n=== FP-Growth Association Rules ===")
print(rules_fpgrowth[['antecedents', 'consequents', 'support', 'confidence', 'lift']])

# Efficiency comparison
print("\n=== Efficiency Comparison ===")
print(f"Apriori Execution Time: {apriori_time:.6f} seconds")
print(f"FP-Growth Execution Time: {fpgrowth_time:.6f}seconds")
```
